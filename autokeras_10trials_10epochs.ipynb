```python
import autokeras as ak
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
import time
import os

# Load mnist dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Initialize the image classifier
clf = ak.ImageClassifier(max_trials=10)

# Record start time
start_time = time.time()

# Fit the image classifier with training data
clf.fit(x_train, y_train, epochs=10)

# Calculate and print training duration
end_time = time.time()
training_duration = end_time - start_time
print(f"Training Time: {training_duration:.2f} seconds")

# Extract the results of all trials
all_trials = clf.tuner.oracle.get_best_trials(num_trials=10)

# Extract and plot validation accuracy for each trial
val_accuracies = [trial.metrics['val_accuracy'].get_best_value() for trial in all_trials]
trial_ids = [trial.trial_id for trial in all_trials]

# Plot validation accuracies
plt.figure(figsize=(12, 6))
plt.bar(trial_ids, val_accuracies, color='blue')
plt.xlabel('Trial ID')
plt.ylabel('Validation Accuracy')
plt.title('Validation Accuracy for Each Trial')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Evaluate the best model with testing data
print("Best Model Evaluation:")
print(clf.evaluate(x_test, y_test))

# Export as a keras model
model = clf.export_model()

# Convert to tensorflowjs model
import tensorflowjs as tfjs
tfjs.converters.save_keras_model(model, 'tfjs_model')

# Save the model
model_path = 'model.h5'
model.save(model_path)

# Print model size
model_size = os.path.getsize(model_path)
print(f"Model Size: {model_size / (1024 * 1024):.2f} MB")

# Print the number of trainable parameters
print(f"Number of Trainable Parameters: {model.count_params()}")

# Load the model
from tensorflow.keras.models import load_model
model = load_model(model_path)

# Predict with the model
import numpy as np
img = np.expand_dims(x_test[0], axis=0)
predicted_y = model.predict(img)
print(predicted_y)
```
